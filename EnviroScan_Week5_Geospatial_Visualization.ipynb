```python
import folium
```

```python
map_center = [
    df['Latitude'].mean(),
    df['Longitude'].mean()
]

map_center
```

```python
base_map = folium.Map(
    location=map_center,
    zoom_start=2,
    tiles='cartodbpositron'
)
```

```python
base_map
```

```python
source_colors = {
    'Industrial': 'red',
    'Vehicular': 'blue',
    'Agricultural': 'green',
    'Natural': 'gray'
}
```

```python
sample_df = df.sample(1000, random_state=42)
```

```python
for _, row in sample_df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=4,
        color=source_colors.get(row['pollution_source'], 'black'),
        fill=True,
        fill_opacity=0.6,
        popup=(
            f"City: {row['City']}<br>"
            f"Country: {row['Country']}<br>"
            f"AQI: {row['AQI Value']}<br>"
            f"Source: {row['pollution_source']}"
        )
    ).add_to(base_map)
```

```python
base_map
```

```python
from folium.plugins import HeatMap
```

```python
heat_data = sample_df[
    ['Latitude', 'Longitude', 'AQI Value']
].values.tolist()
```

```python
HeatMap(
    heat_data,
    radius=15,
    blur=20,
    max_zoom=5
).add_to(base_map)
```
