# Pollution & Weather Dataset - EDA Notebook
This notebook performs Exploratory Data Analysis on the uploaded dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv("F:\\Internship\\infosys-springboard\\resources\\Pollution_Weather_datset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['AQI Value'], bins=30)
plt.title('Distribution of AQI Value')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Temperature (C)'], bins=30)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['AQI Category'].value_counts().plot(kind='bar', figsize=(8,5))
plt.title('AQI Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
plt.figure(figsize=(12,8))
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.title('Correlation Heatmap')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Temperature (C)'], df['AQI Value'])
plt.title('Temperature vs AQI')
plt.xlabel('Temperature (C)')
plt.ylabel('AQI Value')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Humidity (%)'], df['AQI Value'])
plt.title('Humidity vs AQI')
plt.xlabel('Humidity (%)')
plt.ylabel('AQI Value')
plt.show()
```

```python
top_polluted = df.groupby('City')['AQI Value'].mean().sort_values(ascending=False).head(10)
top_polluted.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Most Polluted Cities')
plt.ylabel('Average AQI')
plt.show()
```

```python
cleanest = df.groupby('City')['AQI Value'].mean().sort_values().head(10)
cleanest.plot(kind='bar', figsize=(10,5), color='green')
plt.title('Top 10 Cleanest Cities')
plt.ylabel('Average AQI')
plt.show()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
df['Date'] = df['Timestamp'].dt.date
df.head()
```

```python
daily_trend = df.groupby('Date')['AQI Value'].mean()
plt.figure(figsize=(12,5))
daily_trend.plot()
plt.title('Daily AQI Trend')
plt.xlabel('Date')
plt.ylabel('Average AQI')
plt.show()
```

```python
pollutants = ['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']
df[pollutants].mean().plot(kind='bar', figsize=(8,5))
plt.title('Average Pollutant AQI Levels')
plt.ylabel('AQI Value')
plt.show()
```
