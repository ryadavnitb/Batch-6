## Outliers detection and elimination

```python
outliers = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
# Example: number of outliers per feature
for col, data in outliers.items():
    print(f"{col}: {len(data)} outliers")
```

```python
clip_99 = ['PM2_5', 'PM10', 'O3', 'CO', 'Wind_Speed']
clip_95 = ['NO2', 'SO2']

# Function to clip columns based on percentile
def clip_columns(df, cols, lower_percentile=0.01, upper_percentile=0.99):
    for col in cols:
        lower = df[col].quantile(lower_percentile)
        upper = df[col].quantile(upper_percentile)
        df[col] = df[col].clip(lower, upper)
    return df
```

```python

# Clip 99th percentile columns
df = clip_columns(df, clip_99, lower_percentile=0.01, upper_percentile=0.99)

# Clip 95th percentile columns
df = clip_columns(df, clip_95, lower_percentile=0.05, upper_percentile=0.95)
```

## Feature engineering

```python
df.columns
```

```python
# Drop exact duplicates
df = df.drop_duplicates()

# Check for invalid records (e.g., negative distances, negative pollutants)
df = df[(df['Latitude'].between(-90, 90)) & (df['Longitude'].between(-180, 180))]
pollutants = ['PM2_5', 'PM10', 'NO2', 'SO2', 'O3', 'CO', 'AQI']
for col in pollutants:
    df = df[df[col] >= 0]  # pollutants can't be negative
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
# Extract temporal features
df['Hour'] = df['Timestamp'].dt.hour
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
df['Month'] = df['Timestamp'].dt.month
```

```python
def assign_source_final_logic(row):
    # 1. INDUSTRIAL: Check this FIRST to protect it
    #  Near Industry + Any Sulfur/NO2 presence
    if (row.Dist_Industry <= 3000 and (row.SO2 > 5 or row.NO2 > 30)):
        return 'Industrial'

    # 2. BURNING: 
    # Must have HIGH CO (Smoke) to distinguish from normal dust
    elif (
        (row.Dist_Dump <= 4000 or row.Dist_Farm <= 4000) and
        (row.PM2_5 > 80 or row.CO > 550) and  # RAISED THRESHOLDS
        (row.Season != 'monsoon')
    ):
        return 'Burning'

    # 3. VEHICULAR: Traffic
    elif (row.Dist_Road <= 1500 and (row.NO2 > 20 or row.CO > 400)):
        return 'Vehicular'

    # 4. AGRICULTURAL: The "Dust" 
    #  Near Farm + High PM10 + LOWER CO (Not burning, just dust)
    elif (
        row.Dist_Farm <= 5000 and 
        row.Season in ['spring', 'summer'] and 
        row.PM10 > 50
    ):
        return 'Agricultural'

    else:
        return 'Natural'

# Apply the refined logic
df['Pollution_Source'] = df.apply(assign_source_final_logic, axis=1)
```
