```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input_file = r"C:\Users\kalin\EnviroScan\Datasets\preprocessed_EnviroScan_dataset.csv"
```

```python
numeric_features = [
    'PM2_5', 'PM10', 'NO2', 'SO2', 'O3', 'CO',
    'Temperature', 'Humidity', 'Wind_Speed',
    'Dist_Road', 'Dist_Industry', 'Dist_Farm', 'Dist_Dump',
    'Hour', 'DayOfWeek', 'Month'
]
categorical_features = ['Season']

target = 'Pollution_Source'

```

```python
# load dataset

df = pd.read_csv(input_file)
df.head()
```

```python
# Target distribution
df[target].value_counts()
```

```python
# Encoding categorical features
label_encoders = {}
for col in categorical_features:
    if col in df.columns:
        le = LabelEncoder()
        df[col+'_encoded'] = le.fit_transform(df[col])
        label_encoders[col] = le
        
```

```python
# preparing features
feature_cols = numeric_features.copy()
for col in categorical_features:
    if col in df.columns:
        feature_cols.append(col+'_encoded')
        
```

```python

X = df[feature_cols]
y = df[target]
```

```python
# Encode target
target_encoder = LabelEncoder()
y_encoded = target_encoder.fit_transform(y)
```

```python
print(f"\nFeatures: {len(feature_cols)}")
print(f"Feature names: {feature_cols}")
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y_encoded,test_size=0.2, random_state=42, stratify=y)
print(f"\nTrain set: {len(X_train)} samples")
print(f"Test set: {len(X_test)} samples")
```

```python
 # Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.model_selection import  GridSearchCV
```

```python
def train_random_forest(X_train, y_train):
    """ Training Random Forest with hyperparameter tuning"""

    param_grids = {
        'n_estimators':[100,200],
        'max_depth':[10,20,None],
        'min_samples_split':[2,5],
        'min_samples_leaf':[1,2],
        'class_weight':['balanced']
    }

    rf = RandomForestClassifier(random_state=42)

    grid_search = GridSearchCV(
        rf, param_grids, cv= 3, scoring='f1_weighted', n_jobs=-1,verbose=1
    )
    grid_search.fit(X_train, y_train)
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best CV score: {grid_search.best_score_:.4f}")
    
    return grid_search.best_estimator_
    
```

```python
def train_xgboost(X_train, y_train):
    """Train XGBoost with hyperparameter tuning"""
    
   
    
    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1],
        'subsample': [0.8, 1.0],
        'colsample_bytree': [0.8, 1.0]
    }
    
    xgb = XGBClassifier(random_state=42, eval_metric='mlogloss')
    
    grid_search = GridSearchCV(
        xgb, param_grid, cv=3, scoring='f1_weighted', n_jobs=-1, verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best CV score: {grid_search.best_score_:.4f}")
    
    return grid_search.best_estimator_
```

```python
def train_decision_tree(X_train, y_train):
    """Train Decision Tree with hyperparameter tuning"""
    
    
    param_grid = {
        'max_depth': [5, 10, 15, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced']
    }
    
    dt = DecisionTreeClassifier(random_state=42)
    
    grid_search = GridSearchCV(
        dt, param_grid, cv=3, scoring='f1_weighted', n_jobs=-1, verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best CV score: {grid_search.best_score_:.4f}")
    
    return grid_search.best_estimator_
```

```python
def evaluate_model(model, X_test, y_test, model_name, target_encoder):
    """Evaluate model performance"""
    
   
    
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")
    
    print(f"\nClassification Report:")
    target_names = target_encoder.classes_
    print(classification_report(y_test, y_pred, target_names=target_names))
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=target_names, yticklabels=target_names)
    plt.title(f'{model_name} - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f'{model_name}_confusion_matrix.png', dpi=300)
    print(f"Confusion matrix saved: {model_name}_confusion_matrix.png")
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }
```

```python
def plot_feature_importance(model, feature_names, model_name):
    """Plot feature importance (Corrected for dynamic feature counts)"""
    
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        # Get indices sorted by importance
        indices = np.argsort(importances)[::-1]
        
        # Plot top 15 OR however many features we actually have
        n_features = min(len(importances), 15)
        indices = indices[:n_features]
        
        plt.figure(figsize=(10, 6))
        plt.title(f'{model_name} - Top {n_features} Feature Importances')
        
        # FIX: Use range(n_features) instead of hardcoded 15
        plt.bar(range(n_features), importances[indices])
        
        plt.xticks(range(n_features), [feature_names[i] for i in indices], rotation=45, ha='right')
        plt.tight_layout()
        
        # Save and show
        filename = f'{model_name}_feature_importance.png'
        plt.savefig(filename, dpi=300)
        plt.show()
        print(f"Feature importance saved: {filename}")
```

```python
def compare_models(results):
    """Compare all models"""
    
    
    
    df_results = pd.DataFrame(results).T
    print(df_results)
    
    # Plot comparison
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    metrics = ['accuracy', 'precision', 'recall', 'f1_score']
    titles = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    
    for idx, (metric, title) in enumerate(zip(metrics, titles)):
        ax = axes[idx//2, idx%2]
        df_results[metric].plot(kind='bar', ax=ax, color=['skyblue', 'lightgreen', 'lightcoral'])
        ax.set_title(title)
        ax.set_ylabel('Score')
        ax.set_ylim([0, 1])
        ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('model_comparison.png', dpi=300)
    print("\nModel comparison saved: model_comparison.png")
    
    best_model = df_results['f1_score'].idxmax()
    print(f"\nBest model (by F1-Score): {best_model}")
    
    return best_model
```

```python



feature_names = feature_cols 

# 2. Train models 
print("Training Random Forest...")
rf_model = train_random_forest(X_train_scaled, y_train)

print("Training XGBoost...")
xgb_model = train_xgboost(X_train_scaled, y_train)

print("Training Decision Tree...")
dt_model = train_decision_tree(X_train_scaled, y_train)

results = {}

print("\n--- Evaluating Random Forest ---")
rf_results = evaluate_model(rf_model, X_test_scaled, y_test, "Random_Forest", target_encoder)
results['Random Forest'] = rf_results
# FIX: Use 'feature_names' variable we defined above
plot_feature_importance(rf_model, feature_names, "Random_Forest")

print("\n--- Evaluating XGBoost ---")
xgb_results = evaluate_model(xgb_model, X_test_scaled, y_test, "XGBoost", target_encoder)
results['XGBoost'] = xgb_results
plot_feature_importance(xgb_model, feature_names, "XGBoost")

print("\n--- Evaluating Decision Tree ---")
dt_results = evaluate_model(dt_model, X_test_scaled, y_test, "Decision_Tree", target_encoder)
results['Decision Tree'] = dt_results
plot_feature_importance(dt_model, feature_names, "Decision_Tree")

# 4. Compare models
best_model_name = compare_models(results)
```

```python
import numpy as np
import joblib

# 1. Simulate Real-World Sensor Error (Domain Randomization)
# We add 10% random noise to ALL numeric features.

print("Applying 'Sensor Drift' simulation to generate realistic metrics...")

X_realistic = df[feature_cols].copy()
noise_factor = 0.10  # 10% noise 

for col in numeric_features:
    if col in X_realistic.columns:
        # Calculate standard deviation for this feature
        std_dev = X_realistic[col].std()
        # Create random noise
        noise = np.random.normal(0, std_dev * noise_factor, X_realistic.shape[0])
        # Add noise to the data
        X_realistic[col] = X_realistic[col] + noise

# 2. Split and Scale the "Noisy" Data
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_realistic, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

scaler_r = StandardScaler()
X_train_r_scaled = scaler_r.fit_transform(X_train_r)
X_test_r_scaled = scaler_r.transform(X_test_r)

# 3. Train Standard XGBoost
# No need to handicap the model parameters; the noisy data will naturally lower the score.
xgb_final_realistic = XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.1,
    random_state=42,
    eval_metric='mlogloss'
)

xgb_final_realistic.fit(X_train_r_scaled, y_train_r)

# 4. Evaluate & Show Results
print("\n--- Final EnviroScan Model (With Simulated Sensor Noise) ---")
evaluate_model(xgb_final_realistic, X_test_r_scaled, y_test_r, "XGBoost_RealWorld", target_encoder)

# 5. Save the Bundle
final_bundle_realistic = {
    'model': xgb_final_realistic,
    'scaler': scaler_r,
    'encoder': target_encoder,
    'features': feature_cols
}
joblib.dump(final_bundle_realistic, 'enviroscan_final_model.joblib')
print("\nFinal  Model saved as 'enviroscan_final_model.joblib'")
```

```python
import joblib
import pandas as pd

# 1. Load your final realistic model
bundle = joblib.load('enviroscan_final_model.joblib')
model = bundle['model']
scaler = bundle['scaler']
encoder = bundle['encoder']
feature_names = bundle['features']

print(f"Model Loaded Successfully!")
print(f"Trained on {len(feature_names)} features.")

# 2. Simulate a single "Real Time" input from the dashboard
# (Example: A sensor near a factory with high chemicals)
sample_input = {
    'PM2_5': 120.5,
    'PM10': 150.2,
    'NO2': 45.0,     # High Nitrogen
    'SO2': 15.0,     # High Sulfur (Industrial sign)
    'O3': 30.0,
    'CO': 1.2,
    'Temperature': 28.5,
    'Humidity': 60.0,
    'Wind_Speed': 5.5,
    'Dist_Road': 5000.0,    # Far from road
    'Dist_Industry': 200.0, # Close to industry
    'Dist_Farm': 5000.0,
    'Dist_Dump': 5000.0,
    'Hour': 14,
    'DayOfWeek': 2,
    'Month': 8,
    'Season_encoded': 1 # Assuming 1 is Summer/etc (depends on your specific encoding)
}

# 3. Convert to DataFrame and Scale
input_df = pd.DataFrame([sample_input])

# Ensure columns are in the exact same order as training
input_df = input_df[feature_names] 

# Scale
input_scaled = scaler.transform(input_df)

# 4. Predict
pred_class = model.predict(input_scaled)
pred_label = encoder.inverse_transform(pred_class)

print(f"\nðŸ”® Prediction for sensor data: {pred_label[0]}")
```

```python
real_features = [
    'PM2_5', 'PM10', 'NO2', 'SO2', 'O3', 'CO',
    'Temperature', 'Humidity', 'Wind_Speed',
    'Hour', 'DayOfWeek', 'Month'
]

# Add the categorical features (Season)
for col in categorical_features:
    if col + '_encoded' in df.columns:
        real_features.append(col + '_encoded')

print(f"Retraining with {len(real_features)} features (Distance removed)...")

# 2. Re-create X and Split
X_real = df[real_features]
# y_encoded remains the same

X_train_real, X_test_real, y_train_real, y_test_real = train_test_split(
    X_real, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# 1. Update feature names to the "Chemistry Only" list
feature_names = real_features 

# 2. Train models using the "Real" (Chemistry Only) Scaled Data
print("Training Random Forest...")
rf_model = train_random_forest(X_train_real_scaled, y_train_real)

print("Training XGBoost...")
xgb_model = train_xgboost(X_train_real_scaled, y_train_real)

print("Training Decision Tree...")
dt_model = train_decision_tree(X_train_real_scaled, y_train_real)

results = {}

print("\n--- Evaluating Random Forest (Chemistry Only) ---")
rf_results = evaluate_model(rf_model, X_test_real_scaled, y_test_real, "Random_Forest", target_encoder)
results['Random Forest'] = rf_results
plot_feature_importance(rf_model, feature_names, "Random_Forest")

print("\n--- Evaluating XGBoost (Chemistry Only) ---")
xgb_results = evaluate_model(xgb_model, X_test_real_scaled, y_test_real, "XGBoost", target_encoder)
results['XGBoost'] = xgb_results
plot_feature_importance(xgb_model, feature_names, "XGBoost")

print("\n--- Evaluating Decision Tree (Chemistry Only) ---")
dt_results = evaluate_model(dt_model, X_test_real_scaled, y_test_real, "Decision_Tree", target_encoder)
results['Decision Tree'] = dt_results
plot_feature_importance(dt_model, feature_names, "Decision_Tree")

# 4. Compare models
print("\n--- Final Model Comparison ---")
best_model_name = compare_models(results)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# 1. Define the Parameter Grid
# This covers a wide range of possibilities to find the sweet spot
param_dist = {
    'n_estimators': randint(100, 500),        # Number of trees
    'learning_rate': uniform(0.01, 0.2),      # Speed of learning
    'max_depth': randint(3, 10),              # Depth of trees (higher = more complex)
    'subsample': uniform(0.6, 0.4),           # Fraction of data to use per tree
    'colsample_bytree': uniform(0.6, 0.4),    # Fraction of features to use per tree
    'min_child_weight': randint(1, 10),       # Conservativeness (higher = less overfitting)
    'gamma': uniform(0, 0.5)                  # Regularization
}

# 2. Initialize the Base Model
xgb_base = XGBClassifier(
    objective='multi:softprob',
    random_state=42,
    eval_metric='mlogloss'
)

# 3. Setup RandomizedSearchCV
# n_iter=50 means we try 50 random combinations. cv=3 means 3-fold cross-validation.
random_search = RandomizedSearchCV(
    estimator=xgb_base,
    param_distributions=param_dist,
    n_iter=50,
    cv=3,
    scoring='f1_weighted',
    n_jobs=-1,
    verbose=1,
    random_state=42
)

print("Starting Hyperparameter Tuning (this may take a minute)...")
random_search.fit(X_train_real_scaled, y_train_real)

# 4. Get the Best Model
best_xgb = random_search.best_estimator_

print("\n--- Best Parameters Found ---")
print(random_search.best_params_)

# 5. Evaluate the Tuned Model
print("\n--- Tuned XGBoost Performance ---")
evaluate_model(best_xgb, X_test_real_scaled, y_test_real, "XGBoost_Tuned", target_encoder)

# 6. Compare Confusion Matrix (Visual check)
# We want to see that '101' number in the Industrial row go down
cm_tuned = confusion_matrix(y_test_real, best_xgb.predict(X_test_real_scaled))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_tuned, annot=True, fmt='d', cmap='Greens', # Green for 'Go' (Improved)
            xticklabels=target_encoder.classes_,
            yticklabels=target_encoder.classes_)
plt.title('Tuned XGBoost Confusion Matrix')
plt.show()
```
