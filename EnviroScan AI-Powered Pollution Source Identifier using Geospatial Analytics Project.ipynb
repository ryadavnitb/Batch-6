```python
import numpy as np
import pandas as pd

pollutant_cols = ['SO2', 'NO2','PM10','PM2.5']

# Step 1: Replace '-' with NaN
df[pollutant_cols] = df[pollutant_cols].replace('-', np.nan)

# Step 2: Convert to numeric
for col in pollutant_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 3: Fill missing values with median
for col in pollutant_cols:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)  
```

```python
df.isnull().sum()
```

```python
aqi_cols = ['AQI_PM2.5', 'AQI_PM10', 'AQI_SO2', 'AQI_NO2', 'AQI']

for col in aqi_cols:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2']

# Calculate yearly averages for all pollutants
yearly_avg = df.groupby('Year')[pollutants].mean().reset_index()

# Plotting yearly average trends
plt.figure(figsize=(10,6))
for col in pollutants:
    sns.lineplot(data=yearly_avg, x='Year', y=col, marker='o', label=col)

plt.title('Yearly Average Pollutant Trends (2021-2023)')
plt.xlabel('Year')
plt.ylabel('Average Concentration')
plt.legend()
plt.tight_layout()
plt.show()
```

This line plot shows the average yearly pollutant concentration

```python
# 1. Distribution Histograms
fig, ax = plt.subplots(2, 2, figsize=(14, 8))

# Distribution of PM2.5
sns.histplot(df['PM2.5'].dropna(), ax=ax[0,0], kde=True, color='r')
ax[0,0].set_title('Distribution of PM2.5')

# Distribution of PM10
sns.histplot(df['PM10'].dropna(), ax=ax[0,1], kde=True, color='b')
ax[0,1].set_title('Distribution of PM10')

# Distribution of SO2
sns.histplot(df['SO2'].dropna(), ax=ax[1,0], kde=True, color='g')
ax[1,0].set_title('Distribution of SO2')

# Distribution of NO2
sns.histplot(df['NO2'].dropna(), ax=ax[1,1], kde=True, color='orange')
ax[1,1].set_title('Distribution of NO2')

plt.tight_layout()
plt.show()
```

The histograms display the overall spread and concentration of four main pollutants: PM2.5, PM10, SO2, and NO2 across all collected records.This means most areas have good air quality most of the time, but specific locations or periods can experience dangerous pollution episodes that require special attention.

```python
# 1. Distribution Histograms with 3 std deviation lines
fig, ax = plt.subplots(2, 2, figsize=(14, 8))

# PM2.5
sns.histplot(df['PM2.5'].dropna(), ax=ax[0,0], kde=True, color='r')
mean_pm25 = df['PM2.5'].mean()
std_pm25 = df['PM2.5'].std()
ax[0,0].axvline(mean_pm25 + 3*std_pm25, color='k', linestyle='--')
ax[0,0].axvline(mean_pm25 - 3*std_pm25, color='k', linestyle='--')
ax[0,0].set_title('Distribution of PM2.5')

# PM10
sns.histplot(df['PM10'].dropna(), ax=ax[0,1], kde=True, color='b')
mean_pm10 = df['PM10'].mean()
std_pm10 = df['PM10'].std()
ax[0,1].axvline(mean_pm10 + 3*std_pm10, color='k', linestyle='--')
ax[0,1].axvline(mean_pm10 - 3*std_pm10, color='k', linestyle='--')
ax[0,1].set_title('Distribution of PM10')

# SO2
sns.histplot(df['SO2'].dropna(), ax=ax[1,0], kde=True, color='g')
mean_so2 = df['SO2'].mean()
std_so2 = df['SO2'].std()
ax[1,0].axvline(mean_so2 + 3*std_so2, color='k', linestyle='--')
ax[1,0].axvline(mean_so2 - 3*std_so2, color='k', linestyle='--')
ax[1,0].set_title('Distribution of SO2')

# NO2
sns.histplot(df['NO2'].dropna(), ax=ax[1,1], kde=True, color='orange')
mean_no2 = df['NO2'].mean()
std_no2 = df['NO2'].std()
ax[1,1].axvline(mean_no2 + 3*std_no2, color='k', linestyle='--')
ax[1,1].axvline(mean_no2 - 3*std_no2, color='k', linestyle='--')
ax[1,1].set_title('Distribution of NO2')

plt.tight_layout()
plt.show()
```

The dashed lines at ±3 standard deviations highlight the presence of extreme pollution events or outliers, which may represent localized pollution episodes requiring attention. This suggests that while average air quality might be acceptable, spikes in pollution should be monitored and mitigated.

```python
# Plot each pollutant's data points and highlight outliers
plt.figure(figsize=(12, 10))

for i, col in enumerate(pollutants):
    mean = df[col].mean()
    std = df[col].std()
    upper_limit = mean + 3*std
    lower_limit = mean - 3*std
    
    outliers = df[(df[col] > upper_limit) | (df[col] < lower_limit)]
    
    plt.subplot(2, 2, i+1)
    sns.scatterplot(data=df, x=df.index, y=col, label='Data')
    sns.scatterplot(data=outliers, x=outliers.index, y=col, color='red', label='Outliers')
    plt.title(f'{col} Outliers')
    plt.xlabel('Index')
    plt.ylabel(col)
    plt.legend()

plt.tight_layout()
plt.show()
```

The above graphs shows scatter plot of all pollutants that have outliers with red color dots 

```python
# Identify and remove outliers for each pollutant
for col in pollutants:
    mean = df[col].mean()
    std = df[col].std()
    lower_limit = mean - 3 * std
    upper_limit = mean + 3 * std
    # Keep only rows where value is within limits
    df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]

# df now has outliers removed
print(f"Dataset shape after removing outliers: {df.shape}")
```

```python
# Plotting after outlier removal
plt.figure(figsize=(12, 10))
for i, col in enumerate(pollutants):
    plt.subplot(2, 2, i+1)
    sns.scatterplot(data=df, x=df.index, y=col, color='blue')
    plt.title(f'{col} After Outlier Removal')
    plt.xlabel('Index')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

The above graphs shows the pollutant after outliers removal

```python
# Calculate IQR for AQI
Q1 = df['AQI'].quantile(0.25)
Q3 = df['AQI'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier removal bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['AQI'] >= lower_bound) & (df['AQI'] <= upper_bound)]

print(lower_bound)
print(upper_bound)
```

Normalizing the data
StandardScaler transforms your data so that each column you apply it to has a mean of 0 and a standard deviation of 1.
It rescales values using the formula:

The values you got indicate successful normalization:
