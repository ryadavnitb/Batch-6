```python
weather_df['Timestamp'] = pd.to_datetime(
    weather_df['Timestamp'],
    unit='s'
)
```

```python
weather_df[['Timestamp']].head()
```

```python
merged_df = pd.merge(
    weather_df,
    pollution_df,
    on=['Country', 'City'],
    how='left',
    suffixes=('_weather', '_pollution')
)
```

```python
merged_df.shape
```

```python
merged_df.head()
```

```python
cols_to_drop = [col for col in merged_df.columns if col.endswith('_pollution')]
cols_to_drop
```

```python
merged_df.drop(columns=cols_to_drop, inplace=True)
```

```python
merged_df.columns = merged_df.columns.str.replace('_weather', '', regex=False)
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df['pm_ratio'] = (
    merged_df['PM2.5 AQI Value'] /
    (merged_df['PM2.5 AQI Value'] + 1)
)
```

```python
severity_map = {
    'Good': 0,
    'Moderate': 1,
    'Poor': 2,
    'Unhealthy': 3,
    'Severe': 4
}

merged_df['aqi_severity'] = merged_df['AQI Category'].map(severity_map)
```

```python
merged_df['wind_effect'] = pd.cut(
    merged_df['Wind Speed (m/s)'],
    bins=[0, 2, 5, 100],
    labels=['Low', 'Medium', 'High']
)
```

```python
merged_df['hour'] = merged_df['Timestamp'].dt.hour
merged_df['month'] = merged_df['Timestamp'].dt.month
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

merged_df['season'] = merged_df['month'].apply(get_season)
```
